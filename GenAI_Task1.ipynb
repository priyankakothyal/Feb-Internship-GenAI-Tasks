**1. Importing Libraries**

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
```

**2. Multi-Input Prompt System**

```python
multi_template = "Explain {topic} for {audience} in a {tone} tone"

multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_template
)

# Test Cases
print(multi_prompt.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_prompt.format(topic="Python", audience="kids", tone="fun"))
print(multi_prompt.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

The above outputs demonstrate how a single reusable template can generate different prompts dynamically based on varying inputs.

**3. Prompt Variations Engine**

```python
# Teaching
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

This task demonstrates how multiple PromptTemplates can be used to generate different styles of prompts for the same input, enabling flexible LLM interactions.

**4. ChatPromptTemplate System**



```python
def get_chat_prompt(role, topic):

    if role == "teacher":
        system_msg = "You are a helpful teacher."
    elif role == "interviewer":
        system_msg = "You are a strict interviewer."
    elif role == "motivator":
        system_msg = "You motivate and inspire learners."
    else:
        system_msg = "You are an assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Explain {topic}")
    ])

    return chat_template.format_messages(topic=topic)


# Test
messages = get_chat_prompt("teacher", "Neural Networks")
for msg in messages:
    print(msg.content)
```

This task demonstrates how ChatPromptTemplate structures prompts into system and user roles, enabling better control over LLM behavior in conversational settings.

**5. Input Validation Layer**

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Setting default: beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Setting default: casual")
        tone = "casual"

    return audience, tone

# Test 1: Valid input
result1 = validate_inputs("beginner", "fun")
print("Test 1 Output:", result1)

# Test 2: Invalid input
result2 = validate_inputs("kids", "cool")
print("Test 2 Output:", result2)
```

This function validates user inputs for audience and tone.
If invalid values are provided, default values are assigned to ensure robustness of the prompt system.

**6. Prompt Generator App**

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} tone step by step"
    elif style == "interview":
        template = "Ask interview questions about {topic} for {audience} in a {tone} tone"
    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"
    else:
        template = "Explain {topic} for {audience} in a {tone} tone"

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt.format(topic=topic, audience=audience, tone=tone)


# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

This task integrates validation and prompt templates to generate dynamic prompts based on user inputs and selected styles, forming a complete prompt generation pipeline.

**7. Template Reusability Test**

```python
reuse_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

test_cases = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "formal"),
    ("Data Science", "expert", "casual"),
    ("ML", "beginner", "formal"),
    ("DL", "expert", "fun")
]

for t in test_cases:
    print(reuse_template.format(topic=t[0], audience=t[1], tone=t[2]))
```

This task demonstrates template reusability by using a single PromptTemplate to generate multiple outputs with different inputs, ensuring scalability and modular design.
